```python
#column name A is same in both dataframe so it will throw an error
df1.join(df2)
```

```python
#to overcome the above error
#left
df1.join(df2,lsuffix='_1')
```

```python
#right
df1.join(df2,rsuffix='_1')
```

# Pivot Table

```python
#to read csv file
df=pd.read_csv("C:\\Users\\charu\\Desktop\\MCC\\5. FIFTH SEMESTER\\MACHINE LEARNING word\\student.csv")
df
```

```python
#it will give the average of group of values
df.pivot_table(index='Department')
```

```python
#it will give the average of group of values columnwise for attendance
df.pivot_table(index='Department',columns='Attendance')
```

```python
#it will give the count of group of values columnwise for attendance
df.pivot_table(index='Department',columns='Attendance',aggfunc='count')
```

```python
#it will give the sum of group of values columnwise for attendance
df.pivot_table(index='Department',columns='Attendance',aggfunc='sum')
```

```python
#it will fill the NaN values with none
df.pivot_table(index='Department',columns='Attendance',fill_value='none')
```

```python
#it will fill the NaN values with 0
df.pivot_table(index='Department',columns='Attendance',fill_value=0)
```

```python
#it will subtotal or grand total all row/column
df.pivot_table(index='Department',columns='Attendance',fill_value=0,margins=True)
```

```python
df.pivot_table(index='Department',columns='Attendance',aggfunc='sum',margins=True,fill_value=0)
```

# Melt()

```python
#it will show us all the values.
pd.melt(df)
```

```python
pd.melt(df,id_vars=['CA1','CA2'])
```

```python
#value_vars is used to unpivot the data
pd.melt(df,id_vars=['CA1'],value_vars=['Department'])
```

```python
#var_name is used to change the label variable
pd.melt(df,id_vars=['CA1'],value_vars=['Department'],var_name='Category')
```

```python
#value_name is used to label the values.
pd.melt(df,id_vars=['CA1'],value_vars=['Department'],var_name='Category',value_name='Department')
```
