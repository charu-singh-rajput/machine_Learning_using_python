```python
#to replace a certain value
df.replace(to_replace='M.C.A.', value='Master of Computer Application')
```

```python
#to replace directly
df.replace('M.Com','Master of Commerce')
```

```python
#to replace more than one values with one value
df.replace([1,2,3,4],80)
```

```python
#replace multiple valules with multiple values
df.replace([1,2,3,4],[80,24,56,3])
```

```python
#for a particular column
df.replace({'Attendance':'98%'},'100%')
```

```python
#more than one column
df.replace({'Attendance':'98%','CA1':87.0},'Not Available')
```

```python
#replacing all the strings vaulues to integer value 0 using regular expression
df.replace('[A-Za-z]',0,regex=True)
```

```python
#for a particular row
df.replace({'Department':'[A-Za-z]'},0,regex=True)
```

```python
#forward fill
df.replace(87.0,method='ffill')
```

```python
#backward fill
df.replace(87.0,method='bfill')
```

```python
#limit is used to fill only that many number of values
df.replace(87.0,method='bfill',limit=1)
```

```python
#will modify the same table
#changing 78.0 to 90.0
df.replace(78.0,90.0,inplace=True)
```

# Pandas interpolate()

```python
#fill the empty values by guessing what answer can come
#guess only numeric values but not string values
df.interpolate()
```

```python
df.interpolate(method='linear')
```

```python
#missing values will be filled according to index values
df.interpolate(method='index')
```

```python
#missing values will be filled according to nearest values
df.interpolate(method='nearest')
```

```python
#taking polynomial of the values
df.interpolate(method='polynomial',order=2)
```

```python
df.interpolate(method='spline',order=1)
```

```python
#original data set
df
```

```python
#data type of each column
df.dtypes
```

```python
#fill only one empty value in column
df.interpolate(limit=1)
```

```python
#filling values backward
df.interpolate(limit=1,limit_direction='backward')
```

```python
#filling values forward
df.interpolate(limit=1,limit_direction='forward')
```

```python
df.interpolate(limit=2,limit_direction='forward')
```

```python
#values will be filled from inside
df.interpolate(limit_area='inside')
```

```python
#values will be filled from outside
df.interpolate(limit_area='outside')
```
