```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("tips")
```

```python
#to plot barplot
sns.barplot()
```

```python
sns.barplot(x=df.total_bill)
```

```python
sns.barplot(x=df.day,y=df.total_bill)
plt.show()
```

```python
#hue is a variable in data to map plot aspects to different colors. 
sns.barplot(x=df.day,y=df.total_bill,hue=df.sex)
plt.show()
```

```python
#hue is a variable in data to map plot aspects to different colors. 
sns.barplot(x='day',y='total_bill',hue='sex',data=df)
plt.show()
```

```python
# to change the order
order=["Sun","Thur","Fri","Sat"]
sns.barplot(x='day',y='total_bill',hue='sex',data=df,order=order)
plt.show()
```

```python
hue_order=["Female","Male"]
sns.barplot(x='day',y='total_bill',hue='sex',data=df,hue_order=hue_order)
plt.show()
```

```python
#By default, each bin of a barplot displays the mean value of a variable. 
sns.barplot(x='day',y='total_bill',hue='sex',data=df,estimator=np.mean)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=df,estimator=np.max)
plt.show()
```

```python
#ci Returns a percentile range from an array of values.
#ci is confidence interval
#The confidence interval is then mean +/- z*sigma ,
#where sigma is the estimated standard deviation of your sample mean, given by sigma = s / sqrt(n) , 
#where s is the standard deviation computed from your sample data and n is your sample size.
sns.barplot(x='day',y='total_bill',hue='sex',data=df,ci=12)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=df,ci=20)
plt.show()
```

```python
# Number of bootstraps to use for computing the confidence interval.
sns.barplot(x='day',y='total_bill',hue='sex',data=df,estimator=np.mean,n_boot=2)
plt.show()
```

```python
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=df,**kwargs)
plt.show()
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=df,**kwargs)
plt.show()
```

```python
ax=sns.barplot(x='day',y='total_bill',data=df,alpha=0.9,linestyle="-.",linewidth=3,edgecolor='g')
ax.set(title="Bar Plot For tips ",xlabel="Days",ylabel="Total Bills")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='day',y='total_bill',data=df,alpha=0.9,linestyle="-.",linewidth=3,edgecolor='g')
plt.title("Bar Plot For tips ",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Total Bills",fontsize=15)
plt.show()
```
