# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["AI","ML","DL","BD","NLP"]
class1_student=[34,56,23,45,34]
class2_student=[34,43,12,34,56]
class3_student=[21,23,23,54,31]
```

```python
#plotting the bar chart 
plt.bar(classes,class1_student)
```

```python
#plotting the bar chart  in horizontal
plt.barh(classes,class1_student)
```

```python
#specifying the width of the bar
plt.bar(classes,class1_student,width=0.5)
```

```python
#to align
plt.bar(classes,class1_student,width=0.5,align='edge')
```

```python
#to change the color
plt.bar(classes,class1_student,width=0.5,align='edge',color='r')
```

```python
#to change the visibility 
plt.bar(classes,class1_student,width=0.5,align='edge',color='r',edgecolor='b',linewidth=3,alpha=0.9,linestyle='--',label="Class Students",visible=True)
```

```python
#to increase the size of the bar chart
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.5,align='edge',color='r',edgecolor='b',linewidth=5,alpha=0.9,linestyle='-',label="Class Students",visible=True)
```

```python
#to label the bar chart
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.5,align='edge',color='r',edgecolor='b',linewidth=5,alpha=0.9,linestyle='-',label="Class Students",visible=True)
plt.title("Students List")
plt.xlabel("Subjects")
plt.ylabel("Students")
plt.show()
```

```python
#to increase the size of the bar chart
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.5,align='edge',color='r',edgecolor='b',linewidth=5,alpha=0.9,linestyle='-',label="Class Students",visible=True)
plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to plot two bar chart 
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width=0.2,align='edge',color='r',label="Class Students",visible=True)
plt.bar(classes,class2_student,width=0.2,align='edge',color='y',label="Class Students",visible=True)

plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to plot two bar chart 
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_student,width,align='edge',color='r',label="Class Students",visible=True)
plt.bar(classes_index+width,class2_student,width,align='edge',color='y',label="Class Students",visible=True)

plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to label the x axis values
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_student,width,align='edge',color='r',label="Class Students",visible=True)
plt.bar(classes_index+width,class2_student,width,align='edge',color='y',label="Class Students",visible=True)

plt.xticks(classes_index+width,classes)
plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to rotate the label in x axis values
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_student,width,align='edge',color='r',label="Class Students",visible=True)
plt.bar(classes_index+width,class2_student,width,align='edge',color='y',label="Class Students",visible=True)

plt.xticks(classes_index+width,classes,rotation=90)
plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to rotate the label in x axis values
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_student,width,align='edge',color='r',label="Class Students",visible=True)
plt.bar(classes_index+width,class2_student,width,align='edge',color='y',label="Class Students",visible=True)
plt.bar(classes_index+width+width,class3_student,width,align='edge',color='b',label="Class Students",visible=True)


plt.xticks(classes_index+width,classes)
plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.show()
```

```python
#to rotate the label in x axis values
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.barh(classes_index,class1_student,width,align='edge',color='r',label="Class Students",visible=True)
plt.barh(classes_index+width,class2_student,width,align='edge',color='y',label="Class Students",visible=True)
plt.barh(classes_index+width+width,class3_student,width,align='edge',color='b',label="Class Students",visible=True)


plt.yticks(classes_index+width,classes)
plt.title("Students List",fontsize=25)
plt.ylabel("Subjects",fontsize=25)
plt.xlabel("Students",fontsize=25)
plt.show()
```

```python
#to rotate the label in x axis values
plt.figure(figsize=(16,9))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_student,width,align='edge',color='r',label="First Class",visible=True)
plt.bar(classes_index+width,class2_student,width,align='edge',color='y',label="Second Class",visible=True)
plt.bar(classes_index+width+width,class3_student,width,align='edge',color='b',label="Third Class",visible=True)


plt.xticks(classes_index+width,classes)
plt.title("Students List",fontsize=25)
plt.xlabel("Subjects",fontsize=25)
plt.ylabel("Students",fontsize=25)
plt.legend()
plt.show()
```
