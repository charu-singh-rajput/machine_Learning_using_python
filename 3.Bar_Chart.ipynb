# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["AI","ML","DL","BD","NLP"]
class1_student=[34,56,23,45,34]
class2_student=[34,43,12,34,56]
class3_student=[21,23,23,54,31]
```

```python
#plotting the bar chart 
plt.bar(classes,class1_student)
```

```python
#plotting the bar chart  in horizontal
plt.barh(classes,class1_student)
```

```python
#specifying the width of the bar
plt.bar(classes,class1_student,width=0.5)
```

```python
#to align
plt.bar(classes,class1_student,width=0.5,align='edge')
```

```python
#to change the color
plt.bar(classes,class1_student,width=0.5,align='edge',color='r')
```
