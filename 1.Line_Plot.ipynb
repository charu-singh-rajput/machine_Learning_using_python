# Line Plot

```python
import matplotlib.pyplot as plt
```

```python
#to show the graph between days and temperature
days=[1,2,3,4,5,6]
temperature=[34,34,32.6,23.2,43,34.7]

plt.plot(days,temperature)
plt.show()
```

```python
#to show the graph between days and temperature
days=[1,2,3,4,5,6]
temperature=[34,34,32.6,23.2,43,34.7]

plt.plot(days,temperature)
plt.title('Chennai Temperature')
plt.xlabel('days')
plt.ylabel('temperature')

plt.show()
```

```python
#to change the style
from matplotlib import style
days=[1,2,3,4,5,6]
temperature=[34,34,32.6,23.2,43,34.7]

style.use("ggplot")
plt.plot(days,temperature,'go--',linewidth=2,markersize=10,label="Temperature")
plt.title('Chennai Temperature',fontsize=25)
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=15)
plt.legend(loc=4)
plt.show()
```

```python
#to plot more than one
from matplotlib import style
days=[1,2,3,4,5,6]
chennai_temperature=[34,34,32.6,33.2,43,34.7]
delhi_temperature=[17,24,16.6,63.1,43,84.7]

style.use("ggplot")
plt.plot(days,delhi_temperature,'go--',linewidth=2,markersize=10,label="Delhi Temperature")
plt.plot(days,chennai_temperature,'ko:',linewidth=2,markersize=10,label="Chennai Temperature")
plt.title('Temperature',fontsize=25)
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=15)
plt.grid(color='w',linestyle='-.',linewidth=1)
plt.legend(loc=4)
plt.show()
```
