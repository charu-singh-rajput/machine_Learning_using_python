```python
#creating matrix of range between 0 to 9.42 with the differnce of 0.2 between each element
x=np.arange(0,3*np.pi,0.2)
print(x)
```

```python
#converting all x elments values to sin
y=np.sin(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#converting all x elments values to cos
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
#converting all x elments values to tan
y=np.tan(x)
plt.plot(x,y)
plt.show()
```
