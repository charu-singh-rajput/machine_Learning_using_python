```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# to read the image
img = mpimg.imread('index.png')
print(img)
```

```python
#to find the type of image
type(img)
```

```python
#to find the shape of an array
img.shape
```

```python
#to know the dimension of image
img.ndim
```

```python
#to show the image
plt.imshow(img)
plt.show()
```

```python
#not to show the axis
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
#to increase the size of image
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
#to show the colorbar
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
#to change the colormap
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap='Blues')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap='Accent')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img,cmap='hot')
plt.colorbar()
plt.show()
```

```python
#to get the data for single channel
single_channel=img[:,:,1]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(single_channel,cmap='Blues')
plt.colorbar()
plt.show()
```
