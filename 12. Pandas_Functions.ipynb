```python
#fill all NaN values with 0
df.fillna(0)
```

```python
#fill NaN specifically
df.fillna({'CA1':'None','CA2':'Not Available'})
```

```python
#fill previous values to NaN
df.fillna(method='ffill')
```

```python
#fill backward values to NaN
df.fillna(method='bfill')
```

```python
#fill row wise forward
df.fillna(method='ffill',axis=0)
```

```python
#fill column wise forward
df.fillna(method='ffill',axis=1)
```

```python
#fill row wise backward
df.fillna(method='bfill',axis=0)
```

```python
#fill column wise backward
df.fillna(method='bfill',axis=1)
```

```python
#only one column  vallue will be filled
df.fillna('Not Available',limit=1)
```
