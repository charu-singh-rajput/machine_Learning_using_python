```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import numpy as np
import random
```

```python
#to save the pie chart
plt.pie([40,30,20])
plt.savefig("pie_chart")
plt.show()
```

```python
#to change the resolution for the image
plt.pie([40,30,20])
plt.savefig("pie_chart",dpi=1000)
plt.show()
```

```python
#to change the quality for the image
plt.pie([40,30,20])
plt.savefig("pie_chart_quality",dpi=1000,quality=99)
plt.show()
```

```python
#to give the facecolor 
plt.pie([40,30,20])
plt.savefig("pie_chart_facecolor",dpi=1000,quality=99,facecolor='g')
plt.show()
```
