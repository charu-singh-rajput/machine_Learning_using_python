```python
import matplotlib.pyplot as plt
```

```python
classes=["Python","Java","C++","C","Html"]

class1_student=[56,33,56,34,23]

#plot the pie chart
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_student)
plt.show()
```

```python
#to give label 
plt.pie(class1_student,labels=classes)
plt.show()
```

```python
#to slice 
explode=[0,0,0.2,0,0]
plt.pie(class1_student,labels=classes,explode=explode)
plt.show()
```

```python
explode=[0.2,0,0.2,0,0]
plt.pie(class1_student,labels=classes,explode=explode)
plt.show()
```

```python
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors)
plt.show()
```

```python
#values in float
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.1f%%")
plt.show()
```

```python
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%")
plt.show()
```

```python
#to give shadow
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,explode=explode)
plt.show()
```

```python
#to increase the size 
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,explode=explode,radius=3.4)
plt.show()
```

```python
#to start from 90 degree angle
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,radius=3.4,startangle=90)
plt.show()
```

```python
#to start from 270 degree angle
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,radius=3.4,startangle=270)
plt.show()
```

```python
#to increase the size of the font
textprops={'fontsize':20}
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",explode=[0,0.4,0,0.2,0],shadow=True,radius=2.4,
        startangle=270,textprops=textprops)
plt.show()
```

```python
#to plot pie chart with a good design
wedgeprops={"linewidth":4,'width':1,"edgecolor":'k'}
plt.pie(class1_student,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct="%0.2f%%",
        pctdistance=0.6,
        shadow=True,
        labeldistance=1.4,
        startangle=90,
        radius=1.1,
        counterclock=True,
        wedgeprops=wedgeprops,
        textprops=textprops,
        center=(2,3),
        frame=True,
        rotatelabels=True
       )
```

```python
#to divide each plot into width
wedgeprops={"linewidth":4,'width':2,"edgecolor":'k'}
plt.pie(class1_student,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct="%0.2f%%",
        pctdistance=0.6,
        shadow=True,
        labeldistance=1.4,
        startangle=90,
        radius=1.1,
        counterclock=True,
        wedgeprops=wedgeprops,
        textprops=textprops,
        center=(2,3),
        frame=True,
        rotatelabels=True
       )
```

```python
wedgeprops={"linewidth":4,'width':3,"edgecolor":'k'}
plt.pie(class1_student,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct="%0.2f%%",
        pctdistance=0.6,
        shadow=True,
        labeldistance=1.4,
        startangle=90,
        radius=1.1,
        counterclock=True,
        wedgeprops=wedgeprops,
        textprops=textprops,
        center=(2,3),
        frame=True,
        rotatelabels=True
       )
```

```python
#to increase the size of the font
textprops={'fontsize':20}
colors=['y','g','r','m','c']
plt.pie(class1_student,labels=classes,colors=colors,autopct="%0.2f%%",shadow=True,radius=2.4,
        startangle=270,textprops=textprops)
plt.legend(loc=10)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(8,5))

colors=['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels=np.ones(20)

plt.pie([1],colors='k',radius=2.05)
plt.pie(labels,colors=colors,radius=2.0)

plt.pie([1],colors='g',radius=1.8)
plt.pie([1],colors='y',radius=1.6)
plt.pie([1],colors='c',radius=1.3)
plt.pie([1],colors='b',radius=1.1)
plt.pie([1],colors='m',radius=0.9)

plt.pie([1],colors='b',radius=0.31)
plt.pie(labels,colors=colors,radius=0.3)

plt.pie([1],colors='w',radius=0.2)
plt.pie([1],colors='k',radius=0.1)

plt.show()
```
