```python
#taking only 1000 rows
google_play_store_data = pd.read_csv('C:\\Users\\charu\\Desktop\\MCC\\5. FIFTH SEMESTER\\MACHINE LEARNING word\\googleplaystore.csv',nrows=1000)
google_play_store_data.shape
```

```python
#plotting scatter plot for rating and reviews
x=google_play_store_data["Rating"]
y=google_play_store_data["Reviews"]
plt.scatter(x,y)
```

```python
#giving title and labels to graph
plt.scatter(x,y)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#change the color of the plot
plt.scatter(x,y,c='r')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.scatter(x,y,c='g')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.scatter(x,y,c='k')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#so as to increase the size of the scatter plot
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#to change the plotting mark
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='*')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='+')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='_')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='^')
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#transition gets less visible
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='+',alpha=0.5)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#increases the width of the marker
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='^',linewidths=10)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#to change the color of the plots
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',marker='*',linewidths=10,edgecolors='g',alpha=0.5)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
#to show in verts format
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='k',verts='<',linewidths=10)
plt.title("Google Play Store Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```
