```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#to draw a line plot.
df=pd.DataFrame({"days":[1,2,3,4,5,6],"Temperature":[34.45,25,45,56,43.6,54]})
sns.lineplot(x="days",y="Temperature",data=df)
plt.show()
```

```python
#First search github seaborn dataset in the google search
#Open the first link (https://github.com/mwaskom/seaborn-data)
#There will be many dataset available
#We can use and csv file directly.
df=sns.load_dataset("tips")
df
```

```python
#gives the size of dataset
df.shape
```

```python
#plot the line for x and y
sns.lineplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",data=df)
plt.show()
```

```python
#Map plot aspects to different colors.
sns.lineplot(x="size",y="total_bill",data=df,hue="sex")
plt.show()
```

```python
#Style to line
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex")
plt.show()
```

```python
#Six variations of its default color palette: deep , muted , pastel , bright , dark , and colorblind .
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot")
plt.show()
```

```python
#No dashes will be shown in graph
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=False)
plt.show()
```

```python
#Dashes will be shown in graph
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True)
plt.show()
```

```python
#Mark the plot
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"])
plt.show()
```

```python
#The legend wont be shown
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"],legend=False)
plt.show()
```

```python
#The legend is brief
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"],legend='brief')
plt.show()
```

```python
#To title and label the lineplot
sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"],legend=False)

plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=20)
plt.ylabel("Total Bill",fontsize=20)

plt.show()
```

```python
#To set the backgroud as dark grid
sns.set(style='darkgrid')

sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"],legend=False)

plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=20)
plt.ylabel("Total Bill",fontsize=20)

plt.show()
```

```python
#To change the size of line plot
plt.figure(figsize=(16,9))

sns.set(style='darkgrid')

sns.lineplot(x="size",y="total_bill",data=df,hue="sex",style="sex",palette="hot",dashes=True,markers=["o","<"],legend=False)

plt.title("Line Plot",fontsize=20)
plt.xlabel("Size",fontsize=20)
plt.ylabel("Total Bill",fontsize=20)

plt.show()
```
