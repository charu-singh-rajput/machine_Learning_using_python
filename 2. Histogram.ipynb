# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
CLanguage_student_age=np.random.randint(18,75,(100))
Java_student_age=np.random.randint(12,80,(100))
print(CLanguage_student_age)
print(Java_student_age)
```

```python
#to show the data in the form of histogram
plt.hist(Java_student_age)
plt.show()
```

```python
#to show the data in the form of histogram
plt.hist(Java_student_age)
plt.title("Java Student Age Group")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#to divide the histogram into equal bins
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins)
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#to give a gap between each graph 
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8)
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#to print graph in step
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='step')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#align to left
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',align='left')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#align to right
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',align='right')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#histogram in horizontal 
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',orientation='horizontal')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#to change the color 
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='m')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='r')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='k')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.show()
```

```python
#to mark the bars
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='b',label='Age count')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```

```python
#to present two age group details in same
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='b',label='Java Student')
plt.hist(CLanguage_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='C Student')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```

```python
bins=[10,20,30,40,50,60,70,80]
plt.hist(Java_student_age,bins,rwidth=0.8,histtype='bar',color='b',label='Java Student',orientation='horizontal')
plt.hist(CLanguage_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='C Student')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```

```python
#to show data separately
bins=[10,20,30,40,50,60,70,80]
plt.hist([Java_student_age,CLanguage_student_age],bins,rwidth=0.8,histtype='bar',color=['b','m'],label=['Java Student','C Student'],orientation='horizontal')
#plt.hist(CLanguage_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='C Student')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```

```python
bins=[10,20,30,40,50,60,70,80]
plt.hist([Java_student_age,CLanguage_student_age],bins,rwidth=0.8,histtype='bar',color=['b','m'],label=['Java Student','C Student'])
#plt.hist(CLanguage_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='C Student')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```

```python
#to increase the size of the histogram
bins=[10,20,30,40,50,60,70,80]
plt.figure(figsize=(16,9))
plt.hist([Java_student_age,CLanguage_student_age],bins,rwidth=0.8,histtype='bar',color=['b','m'],label=['Java Student','C Student'])
#plt.hist(CLanguage_student_age,bins,rwidth=0.8,histtype='bar',color='m',label='C Student')
plt.title("Java Student Age Group",fontsize=15)
plt.xlabel("Count",fontsize=15)
plt.ylabel("Age",fontsize=15)
plt.legend()
plt.show()
```
