```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#This module contains a large number of probability distributions as well as a growing library of statistical functions.
#A normal continuous random variable.
from scipy.stats import norm
```

```python
#to represent data in histogram
df=sns.load_dataset("tips")
df
```

```python
sns.distplot(df["size"])
plt.show()
```

```python
#Plot is distributed in 100 bins
sns.distplot(df["total_bill"],bins=100)
plt.show()
```

```python
#The curve line will only be ploted that is KDE(Kernel Density Estimation)
#Is a way to estimate the probability density function of a continuous random variable.
sns.distplot(df["total_bill"],hist=False)
plt.show()
```

```python
#KDE will not be represented
sns.distplot(df["total_bill"],kde=False)
plt.show()
```

```python
#A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. 
#It is used to visualise the distribution of the data.
sns.distplot(df["total_bill"],rug=True)
plt.show()
```

```python
#It will plot a normal linw  which represents a continous values
sns.distplot(df["total_bill"],fit=norm)
plt.show()
```

```python
sns.distplot(df["total_bill"],kde=False,fit=norm)
plt.show()
```

```python
#To change the color of the plot
sns.distplot(df["total_bill"],color='r')
plt.show()
```

```python
#Plot in vertical
sns.distplot(df["total_bill"],vertical=True)
plt.show()
```

```python
#Plot a normal histogram
sns.distplot(df["total_bill"],norm_hist=True)
plt.show()
```

```python
#To change the x-axis label name
sns.distplot(df["total_bill"],axlabel="Total Bill")
plt.show()
```

```python
#To show the legend
sns.distplot(df["total_bill"],label="Total Bill")
plt.legend()
plt.show()
```

```python
#To give the title
sns.distplot(df["total_bill"],label="Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
#to increase the size of image
plt.figure(figsize=(16,9))
sns.distplot(df["total_bill"],label="Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
#To give proper spacing between each plot
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(df["total_bill"],label="Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
#The total_bill column values will be sorted
df.total_bill.sort_values()
```

```python
#The labels of x-axis have been changed and distributed according to bins
bins=[1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(16,9))
sns.set()
sns.distplot(df["total_bill"],label="Total Bill",bins=bins)

plt.xticks(bins)

plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```
