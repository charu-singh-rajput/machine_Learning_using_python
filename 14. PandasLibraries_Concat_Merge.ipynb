# Merging()


```python
#creating dataframe using dictionary
df1=pd.DataFrame({'ID':[1,2,3,4,5],
                 'Class':[45,23,21,34,22]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5],
                 'Name':['Charu','Sakthi','Lenora','Rose','Ammy']})
df2
```

```python
#merging two data frame files
pd.merge(df1,df2)
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2,on='ID')
```

```python
df3=pd.DataFrame({'ID':[1,2,8,4,7],
                 'Class':[45,23,21,34,22]})
df3
```

```python
df4=pd.DataFrame({'ID':[1,2,3,4,5],
                 'Name':['Charu','Sakthi','Lenora','Rose','Ammy']})
df4
```

```python
pd.merge(df3,df4)
```

```python
#print all according to left table
pd.merge(df3,df4,how='left')
```

```python
#print only common
pd.merge(df3,df4,how='inner')
```

```python
#print all according to right table
pd.merge(df3,df4,how='right')
```

```python
#print all
pd.merge(df3,df4,how='outer')
```

```python
#will tell us whetther values are in both the tables or not
pd.merge(df3,df4,how='outer',indicator=True)
```

```python
#to show all the values of both the tables
pd.merge(df3,df4,left_index=True,right_index=True)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['Charu','Sakthi','Lenora','Rose']})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                 'Name':['Manu','Vayuna','Priyanka','Sunaina']})
df2
```

```python
#concatenating two dataframe of same type
#index values are repeating
pd.concat([df1,df2])
```

```python
#concatenating two dataframe of same type
#index values are not repeating
pd.concat([df1,df2],ignore_index=True)
```

```python
#concatenating two dataframe of same type columnwise
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                 'Name':['Charu','Sakthi','Lenora','Rose']})
df1
```

```python
df2=pd.DataFrame({'ID':[5,6],
                 'Name':['Manu','Vayuna']})
df2
```

```python

pd.concat([df1,df2])
```

```python
#concatenating two dataframe columnwise with different number of elements in it
pd.concat([df1,df2],axis=1)
```

```python
#using inner join we can avoid the NaN values
pd.concat([df1,df2],axis=1,join='inner')
```

```python
#based n df1 no of index values are printed
pd.concat([df1,df2],axis=1,join_axes=[df1.index])
```

```python
#based n df2 no of index values are printed
pd.concat([df1,df2],axis=1,join_axes=[df2.index])
```

```python
#used to label the dataframe 
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
#used to label the dataframe in column
pd.concat([df1,df2],keys=['df1','df2'],axis=1)
```
